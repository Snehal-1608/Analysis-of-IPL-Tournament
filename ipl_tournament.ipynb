# what is the probability of winning team if toss is also winning

```python
df[["toss_winner","winner"]]
```

```python
won=0
lost=0
for ind in df.index:
    if df["toss_winner"][ind]==df["winner"][ind]:
        won=won+1
    else:
        lost=lost+1
```

```python
plt.figure(figsize=(8,6))
plt.title("Won vs Lost after winning toss")
plt.pie((lost,won), labels=("Lost the match","Won the match"),autopct="%1.2f%%",startangle=90)
```
